```python
import pandas as pd

df = pd.read_json('data.json')

```

```python
# How many cities (including subregions) where Swiggy is having its restaurants listed?
city_columns = df.columns[1:]  
all_cities = df[city_columns].values.flatten()
```

```python
# unique_cities = set(all_cities)

# Get the count of unique cities (including subregions) where Swiggy has its restaurants listed
total_cities = len(all_cities)

print(f"Total cities (including subregions) with Swiggy restaurants listed: {total_cities}")
```

```python
# How many cities  (don't include subregions) where Swiggy is having their restaurants listed?
# Filter out subregions from the city list (remove 'NaN' and None values)
cities_without_subregions = [city for city in all_cities if city is not None and str(city) != 'NaN' and 'link' not in str(city)]
# cities_without_subregions = set(cities_without_subregions)
total= len(cities_without_subregions)
print(f"Total cities (excluding subregions) with Swiggy restaurants listed: {total}")
```

```python
# The Subregion of Delhi with the maximum number of restaurants listed on Swiggy?

delhi_df = df[df.apply(lambda row: 'Delhi' in row.name, axis=1)]
subregion_counts = delhi_df.count()
max_subregion = subregion_counts.idxmax()

print("Subregion of Delhi with the maximum number of restaurants listed on Swiggy:", max_subregion)
```

```python
# Name the top 5 Most Expensive Cities in the Datasets.

restaurant_counts = df.count()
sorted_cities = restaurant_counts.sort_values(ascending=False)
top_5_most_expensive_cities = sorted_cities.index[:5]

print("Top 5 Most Expensive Cities:")
for city in top_5_most_expensive_cities:
    print(city)
```

```python
# List out the top 5 Restaurants with Maximum & minimum ratings throughout the dataset.
restaurant_df = pd.DataFrame()
for city_column in df.columns:
    city_restaurants = df[city_column]['restaurants']
    if not pd.isna(city_restaurants):
        for restaurant_id, restaurant_info in city_restaurants.items():
            restaurant_name = restaurant_info['name']
            restaurant_rating = restaurant_info.get('rating', None)
            if restaurant_rating and restaurant_rating not in ['--', 'NA']:
                try:
                    restaurant_rating = float(restaurant_rating)
                except ValueError:
                    restaurant_rating = None
            else:
                restaurant_rating = None
            restaurant_df = restaurant_df.append({
                'Restaurant': restaurant_name,
                'Rating': restaurant_rating
            }, ignore_index=True)

top_5_highest = restaurant_df.dropna().sort_values(by='Rating', ascending=False).head(5)
top_5_lowest = restaurant_df.dropna().sort_values(by='Rating').head(5)

print("Top 5 Restaurants with Maximum Ratings:")
print(top_5_highest)

print("\nTop 5 Restaurants with Minimum Ratings:")
print(top_5_lowest)
```

```python
# Name of top 5 cities with the highest number of restaurants listed.
restaurant_counts = df.count()
highest_restaurants = restaurant_counts.sort_values(ascending=False).head(5)
print("Top 5 Cities with the Highest Number of Restaurants Listed:")
print(highest_restaurants)
```

```python
# Top 10 cities as per the number of restaurants listed?
res_10_cities = restaurant_counts.sort_values(ascending=False).head(10)
print("Top 10 cities as per the number of restaurants listed:")
print(res_10_cities)
```

```python
# Top 5 most popular restaurant chains in India?
import pandas as pd

# Assuming you already have the DataFrame 'df' loaded from the JSON file.

# Step 1: Create a new DataFrame to store restaurant information
restaurant_info_df = pd.DataFrame()

# Step 2: Extract restaurant names and ratings from the 'restaurants' column
for city_column in df.columns:
    city_restaurants = df[city_column]['restaurants']
    if not pd.isna(city_restaurants):
        for restaurant_id, restaurant_info in city_restaurants.items():
            restaurant_name = restaurant_info['name']
            restaurant_rating = restaurant_info.get('rating', None)
            if restaurant_rating and restaurant_rating != '--':
                try:
                    restaurant_rating = float(restaurant_rating)
                except ValueError:
                    restaurant_rating = None
            else:
                restaurant_rating = None
            restaurant_info_df = restaurant_info_df.append({
                'Restaurant': restaurant_name,
                'Rating': restaurant_rating
            }, ignore_index=True)

# Step 3: Drop rows with missing restaurant names
restaurant_info_df.dropna(subset=['Restaurant'], inplace=True)

# Step 4: Group by restaurant names and count the number of listings for each chain
restaurant_counts = restaurant_info_df['Restaurant'].value_counts()

# Step 5: Get the top 5 most popular restaurant chains
top_5_popular_chains = restaurant_counts.head(5)

print("Top 5 Most Popular Restaurant Chains in India:")
print(top_5_popular_chains)
```

```python
# Top 10 Restaurants with Maximum Ratings in Banglore
import pandas as pd

# Assuming you already have the DataFrame 'df' loaded from the JSON file.

# Step 1: Filter rows for Bangalore (Banglore)
bangalore_df = df['Bangalore']

# Step 2: Create a new DataFrame to store restaurant information
restaurant_info_df = pd.DataFrame(columns=['Restaurant', 'Rating'])  # Specify the columns explicitly

# Step 3: Check if the DataFrame is not empty before extracting restaurant names and ratings for Bangalore
if not bangalore_df.empty and not pd.isna(bangalore_df['restaurants']):
    for restaurant_id, restaurant_info in bangalore_df['restaurants'].items():
        restaurant_name = restaurant_info['name']
        restaurant_rating = restaurant_info.get('rating', None)
        if restaurant_rating and restaurant_rating != '--':
            try:
                restaurant_rating = float(restaurant_rating)
            except ValueError:
                restaurant_rating = None
        else:
            restaurant_rating = None
        restaurant_info_df = restaurant_info_df.append({
            'Restaurant': restaurant_name,
            'Rating': restaurant_rating
        }, ignore_index=True)

# Step 4: Drop rows with missing ratings
restaurant_info_df.dropna(subset=['Rating'], inplace=True)

# Step 5: Sort restaurants based on ratings in descending order
top_10_highest_rated = restaurant_info_df.sort_values(by='Rating', ascending=False).head(10)

print("Top 10 Restaurants with Maximum Ratings in Bangalore:")
print(top_10_highest_rated)

```

```python
import pandas as pd

# Assuming you already have the DataFrame 'df' loaded from the JSON file.

# Step 1: Filter rows for Bangalore (Banglore)
bangalore_df = df['Bangalore']

# Step 2: Create a new DataFrame to store restaurant information
restaurant_info_df = pd.DataFrame(columns=['Restaurant', 'Rating'])

# Step 3: Check if the DataFrame is not empty before extracting restaurant ratings for Bangalore
if not bangalore_df.empty and not pd.isna(bangalore_df['restaurants']):
    for city_column in bangalore_df.index:
        city_restaurants = bangalore_df.loc[city_column, 'restaurants']
        if not pd.isna(city_restaurants):
            for restaurant_id, restaurant_info in city_restaurants.items():
                restaurant_name = restaurant_info['name']
                restaurant_rating = restaurant_info.get('rating', None)
                if restaurant_rating and restaurant_rating != '--':
                    try:
                        restaurant_rating = float(restaurant_rating)
                    except ValueError:
                        restaurant_rating = None
                else:
                    restaurant_rating = None
                restaurant_info_df = restaurant_info_df.append({
                    'Restaurant': restaurant_name,
                    'Rating': restaurant_rating
                }, ignore_index=True)

# Step 4: Drop rows with missing ratings
restaurant_info_df.dropna(subset=['Rating'], inplace=True)

# Step 5: Sort restaurants based on ratings in descending order
top_10_highest_rated = restaurant_info_df.sort_values(by='Rating', ascending=False).head(10)

print("Top 10 Restaurants with Maximum Ratings in Bangalore:")
print(top_10_highest_rated)
```
